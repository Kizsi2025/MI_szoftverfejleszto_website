```python
import matplotlib.pyplot as plt
import numpy as np

# A Jupyter-ben a grafikon megjelenítéséhez
%matplotlib inline

# Egyszerű adatok generálása
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Grafikon létrehozása
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='sin(x)')
plt.title('Egyszerű szinusz grafikon')
plt.xlabel('X tengely')
plt.ylabel('Y tengely')
plt.legend()
plt.grid(True)
plt.show()

print("A grafikon sikeresen elkészült.")
```

```python
message = "Ez a kurzus a szoftverfejlesztés alapjait tanítja."

# A string átalakítása nagybetűssé
upper_message = message.upper()
print(f"Nagybetűs üzenet: {upper_message}")

# Egy szó cseréje
new_message = message.replace("szoftverfejlesztés", "AI-fejlesztés")
print(f"Módosított üzenet: {new_message}")
```
